```python
## Importing necessary Libraries
import pandas as pd
from datetime import datetime, timedelta
```

```python
## Defining 3 data frames to capture 3 scenarios
df7CDays = pd.DataFrame()
dfLT_10Hrs = pd.DataFrame()
dfMT_14Hrs = pd.DataFrame()
# Load the Excel file
file_path = "Assignment_Timecard.xlsx"
df = pd.read_excel(file_path, engine="openpyxl")

# Convert Time and Time Out columns to datetime objects
df["Time"] = pd.to_datetime(df["Time"], format="%m/%d/%Y %I:%M %p")
df["Time Out"] = pd.to_datetime(df["Time Out"], format="%m/%d/%Y %I:%M %p")

# Sort the DataFrame by Employee Name and Time
df.sort_values(by=["Employee Name", "Time"], inplace=True)

# Initialize variables to track consecutive days and shift duration
consecutive_days = 0
previous_employee = None
previous_time_out = None

# Iterate over the records
for index, row in df.iterrows():
    employee_name = row["Employee Name"]
    time_in = row["Time"]
    time_out = row["Time Out"]
    shift_duration = (time_out - time_in).total_seconds() / 3600  # Convert to hours
    
    # Check if it's the same employee and within the same pay cycle
    if (
        employee_name == previous_employee
        and row["Pay Cycle Start Date"] == previous_row["Pay Cycle Start Date"]
    ):
        # Check for consecutive days
        if (time_in - previous_time_out).total_seconds() / 3600 <= 24:
            consecutive_days += 1
        else:
            consecutive_days = 0
         
        
        # Check for condit_ions (a), (b), and (c)
        if consecutive_days == 7:
            
            df7CDays = df7CDays.append(row)
                     
        elif 1 < (time_in - previous_time_out).total_seconds() / 3600 < 10:
            
            dfLT_10Hrs = dfLT_10Hrs.append(row)
            
        elif shift_duration > 14:
            
            dfMT_14Hrs = dfMT_14Hrs.append(row)

    previous_employee = employee_name
    previous_time_out = time_out
    previous_row = row

```

```python
df7CDays = df7CDays[["Employee Name","Position ID"] ]
dfLT_10Hrs = dfLT_10Hrs[["Employee Name","Position ID"] ]
dfMT_14Hrs = dfMT_14Hrs[["Employee Name","Position ID"] ]
```
